```python
import PyPDF2

def extract_text_from_pdf(pdf_file_path):
    text = ""
    with open(pdf_file_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        for page_num in range(len(pdf_reader.pages)):
            text += pdf_reader.pages[page_num].extract_text()
    return text

pdf_file_path = r"C:\Users\s4021\Downloads\python-exemplars-1.pdf"
pdf_text = extract_text_from_pdf(pdf_file_path)
print(pdf_text)
```

```python
import pdfplumber

def extract_text_from_pdf(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        all_text = ''
        for page in pdf.pages:
            all_text += page.extract_text()
    return all_text

# Example usage
pdf_file_path = r"C:\Users\s4021\Downloads\python-exemplars-1.pdf"
pdf_text = extract_text_from_pdf(pdf_file_path)
print(pdf_text)

```

```python
import speech_recognition as sr

def audio_to_text(audio_file_path):
    # Initialize recognizer
    recognizer = sr.Recognizer()
    
    # Load the audio file
    with sr.AudioFile(audio_file_path) as source:
        audio_data = recognizer.record(source)
    
    # Recognize speech using Google Web Speech API
    try:
        text = recognizer.recognize_google(audio_data)
        return text
    except sr.UnknownValueError:
        return "Google Speech Recognition could not understand audio"
    except sr.RequestError as e:
        return f"Could not request results from Google Speech Recognition service; {e}"

# Example usage
audio_file = r"C:\Users\s4021\Downloads\Recording.wav"
text_output = audio_to_text(audio_file)
print(text_output)

```

```python
import os
from pydub import AudioSegment
from moviepy.editor import VideoFileClip
import speech_recognition as sr

def extract_audio(video_file_path, output_audio_path):
    if os.path.exists(video_file_path):
        video = VideoFileClip(video_file_path)
        audio = video.audio
        audio.write_audiofile(output_audio_path, codec='mp3')
        audio.close()
        video.close()
        print("Audio extracted successfully.")
    else:
        print("Video file not found:", video_file_path)

def convert_mp3_to_wav(mp3_file_path, wav_file_path):
    if os.path.exists(mp3_file_path):
        print("Converting MP3 to WAV...")
        audio = AudioSegment.from_mp3(mp3_file_path)
        audio.export(wav_file_path, format="wav")
        print("Conversion successful.")
    else:
        print("MP3 file not found:", mp3_file_path)

# Set paths for the video and audio files
video_path = r'C:\Users\s4021\OneDrive - Queensland University of Technology\QUT\QUT SEMSTER3\IFN711\2021 S1 IFN711 W2.7 AgileTeams and Techniques v1.mp4'
audio_path = r'C:\Users\s4021\AppData\Local\Programs\Microsoft VS Code\output_audio_file.mp3'
wav_audio_path = r'C:\Users\s4021\AppData\Local\Programs\Microsoft VS Code\output_audio_file.wav'

# Extract audio from video
extract_audio(video_path, audio_path)

# Convert MP3 to WAV
convert_mp3_to_wav(audio_path, wav_audio_path)
```
